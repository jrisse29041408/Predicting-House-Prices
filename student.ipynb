```python
import numpy as np # Just importing all that is needed for EDA
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
kc = pd.read_csv('kc_house_data.csv')
kc.price.describe()
```

```python
kc.info()
```

```python
kc['waterfront'].isna().sum()
```

```python
print(kc['waterfront'].unique())
print(kc['waterfront'].value_counts())
kc['waterfront'] = kc['waterfront'].fillna(kc.waterfront.median())
```

```python
kc['view'].isna().sum()
```

```python
print(kc['view'].unique())
print(kc['view'].value_counts())
kc['view'] = kc['view'].fillna(kc.view.median())
```

```python
kc['yr_renovated'].isna().sum()
```

```python
print(kc['yr_renovated'].unique())
print(kc['yr_renovated'].value_counts())
kc['yr_renovated'] = kc['yr_renovated'].fillna(0.0)
```

```python
kc.info()
```

```python
abs(kc.corr()) > .75
```

```python
ax, fig = plt.subplots(figsize=(15,15))
sns.heatmap(data=kc.corr(), annot=True)
```

```python
# sqft_living15, sqft_above, grade, sqft_living, and bathrooms seem to have a positive correlation on price
sns.set_style('dark')
sns.jointplot(x=kc['sqft_living'], y=kc['price'])
```

```python
sns.jointplot(x=kc['grade'], y=kc['price'])
```

```python
drop = ['zipcode', 'lat', 'long', 'sqft_living', 'sqft_lot15', 'yr_built', 'condition', 'sqft_above', 'id', 'date']
kc = kc.drop(drop, axis=1)
```

```python
kc.info()
kc = kc.drop(['view', 'waterfront'], axis=1)
```

```python
kc.info()
```

```python
kc.grade = kc.grade.astype(str)
kc.floors = kc.floors.astype(str)
kc.bedrooms = kc.bedrooms.astype(str)
kc.bathrooms = kc.bathrooms.astype(str)
cat_kc = pd.DataFrame(data=kc, columns=['grade', 'floors', 'bedrooms', 'bathrooms', 'yr_renovated'])
```

```python
cat_kc = pd.DataFrame(pd.get_dummies(cat_kc, drop_first=True))
```

```python
con_kc = pd.DataFrame(data=kc, columns=['sqft_basement', 'sqft_living15', 'sqft_lot'])
```

```python
cleaned_kc = pd.concat([con_kc, cat_kc], ignore_index=True)
```

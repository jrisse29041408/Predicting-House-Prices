```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('kc_house_data.csv').head(10)
```

```python
df.drop(columns=['id', 'date'], axis=1, inplace=True)
```

```python
df.set_index('grade')
```

```python
df['waterfront'] = df['waterfront'].fillna(df['waterfront'].median())
df['view'] = df['view'].fillna(df['view'].median())
df['yr_renovated'] = df['yr_renovated'].fillna(0)
```

```python
df.loc[df['yr_renovated'] < 2000.0] = 0
df.loc[df['yr_renovated'] >= 2000.0] = 1
df['yr_renovated'] = df['yr_renovated'].astype(int)
df
```
